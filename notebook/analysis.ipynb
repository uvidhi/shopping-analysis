```python
#Import required PYTHON libraries
import zipfile
import os
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Extract the zipfile 
with zipfile.ZipFile("shopping.zip", "r") as zip_ref:
    zip_ref.extractall("shopping_data")
```

```python
# After extracting the ZIP file, list all available CSV files
# to identify the dataset that needs to be loaded.
os.listdir("shopping_data")
```

```python
#Read the dataset
df = pd.read_csv("shopping_data/combined_dataset.csv")
```

```python
#Display first  5 row 
df.head()
```

```python
#Display last 5 row
df.tail()
```

```python
# Check the dimensions of the dataset (rows and columns)
df.shape
```

```python
# Display all column names
df.columns
```

```python
# Check data types and non-null values
df.info()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing values
df["discount"] = df["discount"].fillna(df["discount"].mean())
df["what_customers_said"] = df["what_customers_said"].fillna("Not Available")
df["seller_name"] = df["seller_name"].fillna("Not Available")
df["videos"] = df["videos"].fillna("Not Available")
df["seller_information"] = df["seller_information"].fillna("Not Available")
df["variations"] = df["variations"].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
# Select specific columns for analysis
df[["product_id","title","final_price"]]
```

```python
# Filter records based on conditions
rated_df = df[df["rating"] > 4]
rated_df
```

```python
# Check duplicate records
#  Remove duplicate records if any are present
df.duplicated().sum()
```

```python
# Convert price columns to numeric format
df["final_price"] = (
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
)
df["final_price"]=pd.to_numeric(df["final_price"])
```

```python
#Feature Engineering 
# Create a price difference (savings) column
df["saved_amount"]=df["initial_price"]-df["final_price"]
```

```python
# Create a popularity metric
df["popularity_score"] = df["rating"] * df["ratings_count"]
```

```python
#Category-Level Analysis
# Analyze average rating by category
category_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False)

print(category_rating)
```

```python
# Analyze popularity by category
df["popularity_score"] = df["rating"] * df["ratings_count"]
category_popularity = df.groupby("category")["popularity_score"].mean().sort_values(ascending=False)

print(category_popularity)
```

```python
# Analyze discounts by category
category_savings = df.groupby("category")["saved_amount"].mean().sort_values(ascending=False)

print(category_savings)
```

```python
#Visualization
# Histogram: Distribution of product ratings
plt.hist(df["rating"], bins=10, color="lightgreen", edgecolor="darkgreen")
plt.title("Distribution of product ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

```

```python
#category comparsion
# Bar Chart: Average rating by category
category_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False)

category_rating.head(10).plot(kind="bar",color ="darkgreen")
plt.title("Average Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot: Distribution of final product prices
plt.boxplot(df["final_price"])
plt.title("Distribution of Final Price")
plt.ylabel("Price")
plt.show()
```

```python

sns.heatmap(df[["rating", "ratings_count", "final_price"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
